```python
#FOR PART I
import sys
import numpy as np
from functools import lru_cache,wraps
import collections as col
import string as st
import re
import time
import copy 

sys.setrecursionlimit(15000) 

def ToBaseFive(n):
    if n<5:
        return str(n)
    else:
        return ToBaseFive(int(n/5))+str(n%5)

def ToSnafu(String):
    Snafu=''
    Shift=0
    for char in reversed(list(String)):
        Nb=eval(char)+Shift
        if Nb in [0,1,2]:
            Snafu=str(Nb)+Snafu
            Shift=0
        elif Nb==3:
            Snafu='='+Snafu
            Shift=1
        elif Nb==4:
            Snafu='-'+Snafu
            Shift=1
        elif Nb==5:
            Snafu='0'+Snafu
            Shift=1
    return Snafu

def FromSnafu(String):
    if len(String)==1:
        if String[0] in ['0','1','2']:
            return eval(String[0])
        elif String[0]=='-':
            return -1
        else:
            return -2
    else:
        if String[-1] in ['0','1','2']:
            Nb=eval(String[-1])
        elif String[-1]=='-':
            Nb=-1
        else:
            Nb=-2
        return Nb+5*FromSnafu(String[0:len(String)-1])

def Read_file(file_name):
    with open(file_name) as input_file:
        list_of_lines=input_file.read().splitlines()
    return list_of_lines
            

if __name__ == '__main__':  
    
    List_of_snafu=Read_file('example')
    List_of_snafu=Read_file('input')
    
#     print(List_of_snafu)
    Sum=0
    for snafu in List_of_snafu:
        Number=FromSnafu(snafu)
        print(snafu,Number)
        Sum+=Number
    print(Sum,ToBaseFive(Sum),ToSnafu(ToBaseFive(Sum)))
    
    
```
