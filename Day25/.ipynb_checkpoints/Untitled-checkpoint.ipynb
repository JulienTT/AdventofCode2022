```python
#FOR PART I
import sys
import numpy as np
from functools import lru_cache,wraps
import collections as col
import string as st
import re
import time
import copy 

sys.setrecursionlimit(15000) 

def Read_file(file_name):
    with open(file_name) as input_file:
        list_of_lines=input_file.read().splitlines()
    Robots=[]
    Maps=np.zeros( ( len(list_of_lines),len(list(list_of_lines[0])) ))
    for i in range(1,len(list_of_lines)-1):
        line=list_of_lines[i]
        robots=list(line)
        for j in range(1,len(robots)-1):
            if robots[j]=='>':
                Robots.append([i,j,0,1])
                Maps[i][j]+=1
            elif robots[j]=='<':
                Robots.append([i,j,0,-1])
                Maps[i][j]+=1
            elif robots[j]=='v':
                Robots.append([i,j,1,0])
                Maps[i][j]+=1
            elif robots[j]=='^':
                Robots.append([i,j,-1,0])
                Maps[i][j]+=1
    return Maps,Robots
            
def MoveRobots():
    for robot in Robots:
        y=robot[0]
        x=robot[1]
        Maps[y][x]-=1
        target=[y+robot[2],x+robot[3]]
        if target[0]==0:
            robot[0]=Ly-2
            robot[1]=target[1]
        elif target[0]==Ly-1:
            robot[0]=1
            robot[1]=target[1]
        elif target[1]==0:
            robot[0]=target[0]
            robot[1]=Lx-2
        elif target[1]==Lx-1:
            robot[0]=target[0]
            robot[1]=1
        else:
            robot[0]=target[0]
            robot[1]=target[1]
        Maps[robot[0]][robot[1]]+=1
#     print('Robots have moved')
#     print(Robots)
#     print(Maps)

def MoveUs():
    global Arrived
    global Time
    for pos in list(Paths):
        Paths.discard(pos)
        if (pos[0],pos[1])==(0,1):
            PossibleMoves=[Moves[0],Moves[1]]
        elif (pos[0],pos[1])==(Ly-1,Lx-2):
            PossibleMoves=[Moves[0],Moves[2]]
        else:
            PossibleMoves=Moves
        for move in PossibleMoves:
            newy,newx=pos[0]+move[0],pos[1]+move[1]
            if Maps[newy][newx]==0:
                Paths.add((newy,newx,pos[2]+1))
                if (newy,newx)==Target[currenttarget]:
                    Time+=pos[2]+1
                    print('best time to Target {}'.format(currenttarget),pos[2]+1)
                    Arrived=True
                    break
#     print('time',pos[2]+1,'nb path',len(Paths))

if __name__ == '__main__':  
    
    Maps,Robots=Read_file('example')
    Maps,Robots=Read_file('input')
    Ly=len(Maps)
    Lx=len(Maps[0])    
    
    Target=[(0,1),(Ly-1,Lx-2),(0,1),(Ly-1,Lx-2)]
    currenttarget=0
        
    for i in range(Lx):
        Maps[0][i]=1
        Maps[Ly-1][i]=1
    for i in range(Ly):
        Maps[i][0]=1
        Maps[i][Lx-1]=1
    Maps[0][1]=0
    Maps[Ly-1][Lx-2]=0

    Time=0

    Moves=[[0,0],[1,0],[-1,0],[0,1],[0,-1]]
    for k in range(3):
        Paths=set()
        Arrived=False
        pos=(Target[currenttarget][0],Target[currenttarget][1],0)
        Paths.add(pos)
        currenttarget+=1
        while not Arrived:
            MoveRobots()
            MoveUs()

    print('total Time',Time)
    
        
                    
```
