```python
import sys
import numpy as np
from functools import lru_cache,wraps
import collections as col
import string as st
import re

sys.setrecursionlimit(15000) 

def Read_file(file_name):
    
    with open(file_name) as input_file:
        list_of_lines=input_file.read().splitlines()
    return list_of_lines

def compare(L1,L2):
    #If L2 is empty but not L1, we return True
    if(len(L2)==0 or L2=='[]'):
        if(len(L1)>0 and L1 != '[]'):
            return True
    #In the converse case, we return false
    elif(len(L1)==0 or L1=='[]'):
        if(len(L2)>0 and L2!='[]'):
            return False
    #If L1 is a list
    elif (L1[0]=='['):
        #if L2 is also a list
        if(L2[0]=='['):
            #compare them as list
            compare_list(L1,L2)
        else:
            #otherwise, make L2 a list and compare them as list
            L2='['+L2+']'
            compare_list(L1,L2)
    #if L1 is not empty and not a list, it is a number
    else:
        #if L2 is also a number
        if (L2[0]!='['):
        #compare the numbers
            if(eval(L1[0])<eval(L2[0])):
                return True
            else:
                return False
        #if L2 is a list, make L1 a list and compare them
        else:
            L1='['+L1+']'
            compare_list(L1,L2)
    
    
def compare_list(L1,L2):
    El1=L1.split(',',1)[0]
    El1=El1[1:len(El1)]
    El2=L2.split(',',1)[0]
    El2=El2[1:len(El2)]    
    if(compare(El1,El2)):
        El1='['+re.sub(r'^.*?,', '', L1)
        El2='['+re.sub(r'^.*?,', '', L2)
        return compare(El1,El2)
    
    

if __name__ == '__main__':
    Input=Read_file('example')
    index=1
    Score=0
    L1=[]
    L2=[]
    print(Input)
    for line in Input:
        print(list(line))
        #if there is a packet, read it and store it accordingly
        if (len(line)>0):
            if(len(L1)==0):
                L1=line
            else:
                L2=line
        #between packets, increase index and reset lists
        else:
            L1=[]
            L2=[]
            index+=1
        #If you have read two packets, compare them
        if(L2!=[] and compare(L1,L2)):
            Score+=1
            
            
    print(Input)

```
